```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
fill = np.nanmean(data['NPA.1'])
fill
```

```
data['NPA.1']=data['NPA.1'].fillna(fill)
data['NPA.2']=data['NPA.2'].fillna(fill)
data['NPW.2']=data['NPW.2'].fillna(fill)
data['NPW.1']=data['NPW.1'].fillna(fill)
data['ST4.1']=data['ST4.1'].fillna(fill)
data['ST5.1']=data['ST5.1'].fillna(fill)
data['ST4.2']=data['ST4.2'].fillna(fill)
data['ST5.2']=data['ST5.2'].fillna(fill)
data
```

```
data[data['NPA.1']<0]=np.nan
data[data['NPW.1']<0]=np.nan
data[data['ST4.1']<0]=np.nan
data[data['ST5.1']<0]=np.nan
data[data['NPA.2']<0]=np.nan
data[data['NPW.2']<0]=np.nan
data[data['ST4.2']<0]=np.nan
data[data['ST5.2']<0]=np.nan
```

```
data.describe()
```

```
data['total_no_npa'] = data['NPA.1']*data['NPW.1']
data.describe()
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```

data['Player1'] = le.fit_transform(data['Player1'])
data
```

```
data['Player2'] = le.fit_transform(data['Player2'])
data
```
